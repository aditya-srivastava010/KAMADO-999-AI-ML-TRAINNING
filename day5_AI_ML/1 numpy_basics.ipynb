## 1. Why NumPy?

Python lists are flexible, but they are not optimized for numerical computation.  
NumPy provides the `ndarray`, which is:

- faster for numerical operations,
- memory efficient,
- convenient for vector and matrix calculations,
- heavily used in data science, machine learning, signal processing, and scientific computing.

A NumPy array is usually homogeneous, meaning all elements have the same data type.

````python
# Python list
python_list = [1, 2, 3, 4, 5]

# NumPy array
arr = np.array([1, 2, 3, 4, 5])

print("Python list:", python_list)
print("NumPy array:", arr)
print("Type:", type(arr))
````

## 2. Creating NumPy Arrays

Common ways to create arrays:

- `np.array()`
- `np.zeros()`
- `np.ones()`
- `np.arange()`
- `np.linspace()`
- `np.eye()`

````python
a = np.array([10, 20, 30, 40])
b = np.zeros(5)
c = np.ones((3, 4))
d = np.arange(0, 10, 2)
e = np.linspace(0, 1, 5)
f = np.eye(4)

print("a =", a)
print("\nb =", b)
print("\nc =\n", c)
print("\nd =", d)
print("\ne =", e)
print("\nf =\n", f)
````

## 3. Array Attributes

Important attributes:

- `ndim`: number of dimensions
- `shape`: size along each dimension
- `size`: total number of elements
- `dtype`: data type of elements

````python
x = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Array:\n", x)
print("Number of dimensions:", x.ndim)
print("Shape:", x.shape)
print("Total elements:", x.size)
print("Data type:", x.dtype)
````

## 4. Data Types

NumPy arrays store elements with a fixed data type.  
Common dtypes include:

- `int64`
- `float64`
- `bool`
- `str`

````python
int_arr = np.array([1, 2, 3])
float_arr = np.array([1, 2, 3], dtype=float)

print(int_arr, int_arr.dtype)
print(float_arr, float_arr.dtype)
````

## 5. Indexing and Slicing: 1D Arrays

Indexing is similar to Python lists.

````python
arr = np.array([10, 20, 30, 40, 50, 60])

print("First element:", arr[0])
print("Last element:", arr[-1])
print("Elements from index 1 to 3:", arr[1:4])
print("Every second element:", arr[::2])
print("Reverse array:", arr[::-1])
````

## 6. Indexing and Slicing: 2D Arrays

For 2D arrays:

```python
array[row_index, column_index]
```

````python
matrix = np.array([[10, 20, 30],
                   [40, 50, 60],
                   [70, 80, 90]])

print("Matrix:\n", matrix)

print("\nElement at row 1, column 2:", matrix[1, 2])
print("First row:", matrix[0, :])
print("Second column:", matrix[:, 1])
print("Top-left 2x2 block:\n", matrix[:2, :2])
````

## 7. Reshaping Arrays

`reshape()` changes the shape of an array without changing its data.

Important: the total number of elements must remain the same.

````python
arr = np.arange(1, 13)

reshaped_3x4 = arr.reshape(3, 4)
reshaped_2x6 = arr.reshape(2, 6)

print("Original:", arr)
print("\n3 x 4:\n", reshaped_3x4)
print("\n2 x 6:\n", reshaped_2x6)
````

## 8. Flattening Arrays

Flattening converts a multidimensional array into a 1D array.

````python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("Original matrix:\n", matrix)
print("Flattened:", matrix.flatten())
````

## 9. Vectorized Operations

NumPy avoids explicit loops by applying operations directly to arrays.

This is called **vectorization**.

````python
scores = np.array([45, 50, 65, 80, 90])

# Add 5 marks to every score
new_scores = scores + 5

print("Original scores:", scores)
print("New scores:", new_scores)
````

````python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("b / a =", b / a)
print("a ** 2 =", a ** 2)
````

## 10. Universal Functions

Universal functions, or **ufuncs**, apply mathematical functions element-wise.

````python
arr = np.array([1, 4, 9, 16, 25])

print("Square root:", np.sqrt(arr))
print("Log:", np.log(arr))
print("Exponential:", np.exp([1, 2, 3]))
print("Sine:", np.sin([0, np.pi/2, np.pi]))
````

## 11. Boolean Masking and Filtering

Boolean masks are used to filter arrays based on conditions.

````python
marks = np.array([35, 42, 55, 28, 90, 76, 61])

passing_mask = marks >= 40

print("Marks:", marks)
print("Passing mask:", passing_mask)
print("Passing marks:", marks[passing_mask])
print("Marks greater than 60:", marks[marks > 60])
````

## 12. Multiple Conditions

Use:

- `&` for AND
- `|` for OR
- `~` for NOT

Each condition should be enclosed in parentheses.

````python
marks = np.array([35, 42, 55, 28, 90, 76, 61])

selected = marks[(marks >= 40) & (marks <= 70)]

print("Marks between 40 and 70:", selected)
````

## 13. Broadcasting

Broadcasting allows NumPy to perform operations between arrays with different shapes, when compatible.

Example: adding a scalar to an array.

````python
arr = np.array([10, 20, 30])

print(arr + 5)
````

### Broadcasting with 2D Arrays

A row vector can be added to each row of a matrix.

````python
matrix = np.array([[10, 20, 30],
                   [40, 50, 60],
                   [70, 80, 90]])

row_bonus = np.array([1, 2, 3])

print(matrix + row_bonus)
````

## 14. Basic Statistical Operations

````python
data = np.array([12, 15, 14, 10, 18, 20, 22, 19])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard deviation:", np.std(data))
print("Variance:", np.var(data))
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))
print("Sum:", np.sum(data))
````

## 15. Axis-wise Operations

For a 2D array:

- `axis=0` works column-wise
- `axis=1` works row-wise

````python
marks = np.array([[80, 75, 90],
                  [60, 65, 70],
                  [88, 92, 85]])

print("Marks matrix:\n", marks)

print("\nColumn-wise mean:", np.mean(marks, axis=0))
print("Row-wise mean:", np.mean(marks, axis=1))
````

## 16. Random Number Generation

Use NumPy's random module for simulations, experiments, and reproducible examples.

````python
rng = np.random.default_rng(seed=42)

random_values = rng.random(5)
random_integers = rng.integers(low=1, high=10, size=5)
normal_data = rng.normal(loc=0, scale=1, size=10)

print("Random values:", random_values)
print("Random integers:", random_integers)
print("Normal data:", normal_data)
````

## 17. Matrix Operations

Common matrix operations:

- Matrix addition
- Element-wise multiplication
- Matrix multiplication
- Transpose
- Inverse

````python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("A + B =\n", A + B)
print("\nElement-wise A * B =\n", A * B)
print("\nMatrix multiplication A @ B =\n", A @ B)
print("\nTranspose of A =\n", A.T)
````

### Solving Linear Equations

Example:

\[
2x + y = 8
\]

\[
x + 3y = 13
\]

Matrix form:

\[
AX = b
\]

````python
A = np.array([[2, 1],
              [1, 3]])

b = np.array([8, 13])

solution = np.linalg.solve(A, b)

print("Solution [x, y]:", solution)
````

## 18. Mini-Lab: Student Marks Analysis

We have marks of 5 students in 3 subjects:

- Mathematics
- Physics
- Computer Science

Tasks:

1. Find total marks of each student.
2. Find average marks of each subject.
3. Find the highest total marks.
4. Find students who scored at least 75 average.

````python
marks = np.array([[80, 75, 90],
                  [60, 65, 70],
                  [88, 92, 85],
                  [45, 55, 50],
                  [95, 90, 98]])

students = np.array(["Asha", "Bharat", "Chitra", "Deepak", "Esha"])

total_marks = np.sum(marks, axis=1)
subject_average = np.mean(marks, axis=0)
student_average = np.mean(marks, axis=1)

print("Total marks of each student:")
for name, total in zip(students, total_marks):
    print(name, ":", total)

print("\nAverage marks of each subject:", subject_average)

print("\nHighest total marks:", np.max(total_marks))
print("Student with highest total:", students[np.argmax(total_marks)])

print("\nStudents with average >= 75:")
print(students[student_average >= 75])
````
