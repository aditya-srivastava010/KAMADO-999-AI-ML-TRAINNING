````python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
````

## 1. Why Pandas?

Pandas is used for working with tabular data.

Typical examples:

- student marks table,
- sales data,
- patient records,
- sensor measurements,
- CSV or Excel datasets.

The two most important Pandas objects are:

1. `Series`: one-dimensional labelled data.
2. `DataFrame`: two-dimensional labelled table.

## 2. Pandas Series

A `Series` is like a single column with labels.

````python
marks = pd.Series([80, 75, 90, 65], name="Marks")

print(marks)
print("\nValues:", marks.values)
print("Index:", marks.index)
````

### Series with Custom Index

````python
marks = pd.Series(
    data=[80, 75, 90, 65],
    index=["Asha", "Bharat", "Chitra", "Deepak"],
    name="Marks"
)

print(marks)
print("\nMarks of Chitra:", marks["Chitra"])
````

## 3. DataFrame

A `DataFrame` is like a spreadsheet or SQL table.

````python
data = {
    "Name": ["Asha", "Bharat", "Chitra", "Deepak", "Esha"],
    "Department": ["CSE", "ECE", "CSE", "ME", "ECE"],
    "Math": [80, 60, 88, 45, 95],
    "Physics": [75, 65, 92, 55, 90],
    "Computer": [90, 70, 85, 50, 98]
}

df = pd.DataFrame(data)

df
````

## 4. Inspecting Data

Common inspection methods:

- `head()`
- `tail()`
- `shape`
- `columns`
- `info()`
- `describe()`

````python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())

print("\nFirst rows:")
display(df.head())

print("\nSummary statistics:")
display(df.describe())
````

````python
df.info()
````

## 5. Loading Data

Usually, data is loaded from CSV or Excel files.

For this classroom notebook, we first create a small CSV file and then read it using `pd.read_csv()`.

````python
csv_text = '''Name,Department,Math,Physics,Computer,Attendance
Asha,CSE,80,75,90,92
Bharat,ECE,60,65,70,81
Chitra,CSE,88,92,85,95
Deepak,ME,45,55,50,70
Esha,ECE,95,90,98,96
Farhan,ME,55,,65,75
Gita,CSE,78,82,,88
'''

with open("student_data.csv", "w") as f:
    f.write(csv_text)

students = pd.read_csv("student_data.csv")

students
````

## 6. Selecting Columns

Single column selection returns a Series.

Multiple column selection returns a DataFrame.

````python
# Single column
display(students["Name"])

# Multiple columns
display(students[["Name", "Department", "Math"]])
````

## 7. Selecting Rows with `.iloc`

`.iloc` is integer-position based indexing.

Syntax:

```python
dataframe.iloc[row_positions, column_positions]
```

````python
# First row
display(students.iloc[0])

# First three rows
display(students.iloc[0:3])

# First three rows and first two columns
display(students.iloc[0:3, 0:2])
````

## 8. Selecting Rows with `.loc`

`.loc` is label-based indexing.

It is especially useful with conditions.

````python
# Select rows 0 to 2 and selected columns
display(students.loc[0:2, ["Name", "Department", "Math"]])
````

## 9. Filtering Rows

Filtering is one of the most important operations in Pandas.

````python
# Students with Math marks greater than or equal to 75
high_math = students[students["Math"] >= 75]

high_math
````

### Multiple Conditions

Use:

- `&` for AND
- `|` for OR
- `~` for NOT

Each condition must be inside parentheses.

````python
# CSE students with Math marks >= 75
filtered = students[(students["Department"] == "CSE") & (students["Math"] >= 75)]

filtered
````

## 10. Creating New Columns

Let us compute total marks and average marks.

````python
students["Total"] = students[["Math", "Physics", "Computer"]].sum(axis=1)
students["Average"] = students[["Math", "Physics", "Computer"]].mean(axis=1)

students
````

## 11. Conditional New Column

Let us create a result column:

- Pass if average is at least 40
- Fail otherwise

````python
students["Result"] = np.where(students["Average"] >= 40, "Pass", "Fail")

students[["Name", "Average", "Result"]]
````

## 12. Missing Values

Real datasets often contain missing values.

Important methods:

- `isna()`
- `notna()`
- `fillna()`
- `dropna()`

````python
print("Missing values per column:")
print(students.isna().sum())
````

````python
# Display rows containing at least one missing value
students[students.isna().any(axis=1)]
````

### Filling Missing Values

A common approach is to fill missing numerical values with column mean or median.

````python
students_filled = students.copy()

for col in ["Math", "Physics", "Computer"]:
    students_filled[col] = students_filled[col].fillna(students_filled[col].mean())

students_filled["Total"] = students_filled[["Math", "Physics", "Computer"]].sum(axis=1)
students_filled["Average"] = students_filled[["Math", "Physics", "Computer"]].mean(axis=1)
students_filled["Result"] = np.where(students_filled["Average"] >= 40, "Pass", "Fail")

students_filled
````

## 13. Renaming and Dropping Columns

````python
renamed = students_filled.rename(columns={"Computer": "Computer_Science"})

display(renamed.head())

# Drop a column
without_attendance = renamed.drop(columns=["Attendance"])

display(without_attendance.head())
````

## 14. Sorting Data

````python
# Sort by average marks in descending order
students_sorted = students_filled.sort_values(by="Average", ascending=False)

students_sorted[["Name", "Department", "Average"]]
````

## 15. GroupBy: Category-wise Analysis

`groupby()` is used for split-apply-combine operations.

Example: department-wise average marks.

````python
dept_avg = students_filled.groupby("Department")[["Math", "Physics", "Computer", "Average"]].mean()

dept_avg
````

### Multiple Aggregations

````python
summary = students_filled.groupby("Department").agg(
    Students=("Name", "count"),
    Avg_Math=("Math", "mean"),
    Avg_Physics=("Physics", "mean"),
    Avg_Computer=("Computer", "mean"),
    Avg_Attendance=("Attendance", "mean")
)

summary
````

## 16. Value Counts

`value_counts()` is useful for counting categories.

````python
students_filled["Department"].value_counts()
````

## 17. Merge DataFrames

Merging is similar to joining tables in SQL.

Let us create another DataFrame containing department mentors.

````python
mentors = pd.DataFrame({
    "Department": ["CSE", "ECE", "ME"],
    "Mentor": ["Dr. Sharma", "Dr. Iyer", "Dr. Khan"]
})

merged = pd.merge(students_filled, mentors, on="Department", how="left")

merged[["Name", "Department", "Mentor", "Average"]]
````

## 18. Basic Plotting

Pandas provides simple plotting methods using Matplotlib internally.

For better visualization control, one can use Matplotlib directly.

````python
import matplotlib.pyplot as plt

students_filled.plot(
    x="Name",
    y="Average",
    kind="bar",
    title="Average Marks of Students",
    legend=False
)

plt.ylabel("Average Marks")
plt.xlabel("Student")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

## 19. Mini-Lab: Student Performance Analysis

Tasks:

1. Load the student data.
2. Fill missing values using subject-wise mean.
3. Calculate total and average marks.
4. Assign grade:
   - A: average >= 85
   - B: average >= 70
   - C: average >= 50
   - D: average < 50
5. Find department-wise average performance.
6. Identify the top student.

````python
lab = pd.read_csv("student_data.csv")

# Fill missing values
for col in ["Math", "Physics", "Computer"]:
    lab[col] = lab[col].fillna(lab[col].mean())

# Total and average
lab["Total"] = lab[["Math", "Physics", "Computer"]].sum(axis=1)
lab["Average"] = lab[["Math", "Physics", "Computer"]].mean(axis=1)

# Grade assignment
conditions = [
    lab["Average"] >= 85,
    lab["Average"] >= 70,
    lab["Average"] >= 50
]

choices = ["A", "B", "C"]

lab["Grade"] = np.select(conditions, choices, default="D")

display(lab)

print("Top student:")
display(lab.loc[lab["Average"].idxmax(), ["Name", "Department", "Average", "Grade"]])

print("\nDepartment-wise average:")
display(lab.groupby("Department")["Average"].mean())
````
