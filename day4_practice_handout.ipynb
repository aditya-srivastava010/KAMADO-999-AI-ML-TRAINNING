#***CHALLENGE***

```python
#approach 1
def  grade_report(scores):
  dicti={}
  count=0
  count1=0
  count2=0
  count3=0
  for i in scores:
    if i>=90:
      count+=1
      dicti["A"]=count
    elif i>=80:
      count1+=1
      dicti["B"]=count1
    elif i>=70:
      count2+=1
      dicti["C"]=count2
    else:
      count3+=1
      dicti["F"]=count3

  print(dicti)

#approach2
def  grade_report2(scores):
  dicti={}
  for i in scores:
    if i>=90:
      dicti["A"]=dicti.get("A",0)+1
    elif i>=80:
      dicti["B"]=dicti.get("B",0)+1
    elif i>=70:
      dicti["C"]=dicti.get("C",0)+1
    else:
      dicti["F"]=dicti.get("F",0)+1

  print(dicti)

l1=[95, 82, 71, 68, 90, 55]
grade_report(l1)
grade_report2(l1)
```
