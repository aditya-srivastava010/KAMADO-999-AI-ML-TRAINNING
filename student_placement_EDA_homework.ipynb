# Student Placement EDA


## Step 1: Basic Overview

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_placement_data.csv')
print('Shape:', df.shape)
print('\nColumns:\n', df.columns.tolist())
print('\nData types:\n', df.dtypes)
print('\nNull values:\n', df.isnull().sum())
df.head()
```

## Step 2: Summary Statistics

```python
print('Describe:\n', df.describe())
print('\nPlacement Count:\n', df['Placement Status'].value_counts())
print('\nCoding Skill Count:\n', df['Coding Skill'].value_counts())
```

## Step 3: Placement % Pie Chart

```python
counts = df['Placement Status'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%',
colors=['#1D9E75','#E24B4A'])
plt.title('Placement %')
plt.show()
```

## Step 4: Exam Score Distribution

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='Exam Score', hue='Placement Status', bins=10,
             palette=['#1D9E75','#E24B4A'])
plt.title('Exam Score Distribution')
plt.show()
```

## Step 5: GPA Boxplot

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x='Placement Status', y='GPA',
            palette=['#1D9E75','#E24B4A'])
plt.title('GPA vs Placement')
plt.show()
```

## Step 6: Coding Skill vs Placement

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Coding Skill', hue='Placement Status',
            order=['High','Medium','Low'], palette=['#1D9E75','#E24B4A'])
plt.title('Coding Skill vs Placement')
plt.show()
```

## Step 7: Internship vs Placement

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Internship', hue='Placement Status'
 palette=['#1D9E75','#E24B4A'])
plt.title('Internship vs Placement')
plt.show()
```

## Step 8: College Tier vs Placement

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='College Tier', hue='Placement Status',
              palette=['#1D9E75','#E24B4A'])
plt.title('College Tier vs Placement')
plt.show()
```

## Step 9: IQ vs Exam Score Scatter

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='IQ Score', y='Exam Score',
hue='Placement Status', palette=['#1D9E75','#E24B4A'], s=80)
plt.title('IQ Score vs Exam Score')
plt.show()
```

## Step 10: Correlation Heatmap

```python
plt.figure(figsize=(6,5))
numeric_df = df[['Exam Score','GPA','IQ Score']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
